```python
import pandas as pd
import re

# define a function to remove URLs from a string
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)

# read the CSV file into a pandas DataFrame
df = pd.read_csv('Raw_Customer_reviews.csv')

# drop any rows with missing or null values in the 'Column_1' column
df = df.dropna(subset=['reviews.sourceURLs'])

# apply the remove_urls() function to the 'Column_1' column of the dataframe
df['reviews.sourceURLs'] = df['reviews.sourceURLs'].apply(remove_urls)

# write the updated dataframe to a new CSV file
df.to_csv('Dataset_1.csv', index=False)
```

```python
import pandas as pd
import re

# read in the CSV file
df = pd.read_csv('Dataset_1.csv')

# define the regular expression pattern to match all non-alphanumeric characters
pattern = r'[^a-zA-Z0-9]'

# define a lambda function that applies the pattern to each cell of the dataframe
remove_special_chars = lambda x: re.sub(pattern, '', x) if isinstance(x, str) else x

# apply the lambda function to all cells of the dataframe using the applymap() method
df = df.applymap(remove_special_chars)

# write the cleaned dataframe to a new CSV file
df.to_csv('dataset_2.csv', index=False)
```

```python
import pandas as pd

# read the CSV file into a dataframe
df = pd.read_csv('dataset_2.csv')

# apply the lower() method to a specific column
df['reviews.text'] = df['reviews.text'].str.lower()
df['name'] = df['name'].str.lower()
df['categories']= df['categories'].str.lower()
df['brand'] = df['brand'].str.lower()
df['reviews.title'] = df['reviews.title'].str.lower()
df['reviews.username'] = df['reviews.username'].str.lower()

# write the updated dataframe back to the CSV file
df.to_csv('dataset_3.csv', index=False)

print(df.head(50))
```

```python
import pandas as pd
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

# Load the CSV file into a pandas dataframe
df = pd.read_csv('dataset_3.csv')

# Convert the column of interest to string type
df['reviews.text'] = df['reviews.text'].astype(str)
df['name'] = df['name'].astype(str)
df['categories']= df['categories'].astype(str)
df['brand'] = df['brand'].astype(str)
df['reviews.title'] = df['reviews.title'].astype(str)
df['reviews.username'] = df['reviews.username'].astype(str)

# Tokenize the text in the column of interest

df['reviews.text'] = df['reviews.text'].apply(word_tokenize)
df['name'] = df['name'].apply(word_tokenize)
df['categories']= df['categories'].apply(word_tokenize)
df['brand'] = df['brand'].apply(word_tokenize)
df['reviews.title'] = df['reviews.title'].apply(word_tokenize)
df['reviews.username'] = df['reviews.username'].apply(word_tokenize)


# Save the updated dataframe back to a CSV file
df.to_csv('dataset_4.csv', index=False)

print(df.head(20))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Load your CSV file into a pandas dataframe
df = pd.read_csv('dataset_4.csv')

# Define the stopword list
stop_words = set(stopwords.words('english'))

# Apply the stopword removal on the desired column

df['reviews.text'] = df['reviews.text'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df['name'] = df['name'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df['categories']= df['categories'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df['brand'] = df['brand'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df['reviews.title'] = df['reviews.title'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df['reviews.username'] = df['reviews.username'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df.to_csv('dataset_5.csv', index=False)

print(df.head(10))
```

```python
import pandas as pd
import nltk
from nltk.stem import SnowballStemmer

# Load the CSV file into a pandas dataframe
df = pd.read_csv('dataset_5.csv')

# Initialize the stemmer
stemmer = SnowballStemmer('english')

# Define a function to stem a string of text
def stem_text(text):
    words = nltk.word_tokenize(text)
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Apply the stem_text function to the 'reviews.text' column of the dataframe
df['reviews.text.stemmed'] = df['reviews.text'].apply(stem_text)

# Display a sample of the stemmed data
print(df['reviews.text.stemmed'].head(10))
```

```python
import pandas as pd
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

# read the CSV file into a dataframe
df = pd.read_csv('dataset_5.csv')

# initialize the WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# define a function to lemmatize a list of words
def lemmatize_words(words):
    lemmatized_words = []
    for word in words:
        lemmatized_word = lemmatizer.lemmatize(word)
        lemmatized_words.append(lemmatized_word)
    return lemmatized_words

# lemmatize the text in the column of interest
df['reviews.text'] = df['reviews.text'].apply(lambda x: lemmatize_words(x))

# display a sample of the lemmatized data
print(df['reviews.text'].sample(20))
```
